```python
import csv

class CsvOrm:
    def __init__(self,csv_file):
        self.csv_file = csv_file
        # print(self.csv_file)
        self.data = self._load_data()
        # print(self.data)

    def _load_data(self):
        data = []
        with open(self.csv_file, 'r') as file:
            csv_reader = csv.DictReader(file)
            for row in csv_reader:
                data.append(row)
        return data

    def list(self):
        return self.data

    def get(self, record_id):
        for record in self.data:
            if record['ID'] == record_id:
                return record
        return None

    def filter(self, column_name, value):
        filtered_data = []
        for record in self.data:
            if record.get(column_name) == value:
                filtered_data.append(record)
        return filtered_data


# Initialize the ORM
orm = CsvOrm('data.csv')

# List all records
all_records = orm.list()
print("All records:")
for record in all_records:
    print(record)

# Get a record by ID
id=input("Enter Id:")
record = orm.get(id)
print(f"\nGet record by ID {id}:")
print(record)

# Filter record
age=input("Enter Age:")
filtered_records = orm.filter('Age', age)
print(f"\nFilter records by Age={age}:")
for record in filtered_records:
    print(record)
    
    
# How to create csv file .
import csv

data = [
    ["Name", "Age", "Location"],
    ["John", 30, "New York"],
    ["Jane", 25, "Los Angeles"],
    ["Bob", 35, "Chicago"],
]

with open("data.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)


```
